당화혈색소 => HE_HbA1c, 임신여부 => HE_prg

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/seoihwan/Desktop/escuela/과기대/nyamnaym-jjeobjjeob/data/raw_data/2009년 건강검진조사.csv')
```

```python
df_copy = df.copy()
```

```python
# 20세 이상이면서 임산부가 아닌 데이터
df_na = df[(df.age >= 20) & (df.HE_prg != 1)]
```

```python
df_na
```

```python
df_na = df_na[(df_na.HE_glu >= 126) | (df_na.DE1_dg == 1) | (df_na.DE1_3 == 1) | (df_na.DE1_31 == 1) | (df_na.HE_HbA1c >= 6.5) | (df_na.DE1_32 == 1) | (df_na.DE1_33 == 1) | (df_na.DE1_34 == 1)]
```

```python
df_na
```

```python
df_copy['HE_DM_HbA1c'] = 1
```

```python
df_copy.iloc[df_na.index,-1] = 3
```

```python
df_copy
```

```python
df_2 = df_copy[~df_copy.index.isin(df_na.index)]
df_2 = df_2[(df_2['HE_glu'] >= 100) & (df_2['HE_glu'] <= 125) | (df_2['HE_HbA1c'] >= 5.7) & (df_2['HE_HbA1c'] <= 6.4)]
```

```python
df_copy.iloc[df_2.index,-1] = 2
```
